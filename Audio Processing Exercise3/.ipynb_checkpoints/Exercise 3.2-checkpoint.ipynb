```python
#import libraries for audio analysis
from thinkdsp import read_wave
from thinkdsp import decorate
import thinkdsp
#thinkdsp fractions.gcd(a, b) has been moved to math.gcd(a, b) in Python 3.9.
#Updated the current ver. of thinkdsp.py with the latest ver. on Github
#paramiko maybe causing transport.py errors
```

```python
#Load all sound and assign variable to wave2,wave3,wave4
wave2 = read_wave('audiofiles/Ex3_sound2.wav')
wave3 = read_wave('audiofiles/Ex3_sound3.wav')
wave4 = read_wave('audiofiles/Ex3_sound4.wav')
```

```python
spectrum2 = wave2.make_spectrum()
spectrum3 = wave3.make_spectrum()
spectrum4 = wave4.make_spectrum(full=True)
spectrum2.plot(label = "sound2.wav")
spectrum3.plot(label = "sound3.wav")
spectrum4.plot(label = "sound4.wav")
decorate(xlabel='Frequency (Hz)')
decorate(ylabel='Amplitude')
```

This is what the spectrum looks like with response to Amplitude and Frequency, As shown in the graph above, there is a spike in **sound4.wav** in the high frequency range **around 20kHz.** Human auditory system can pick up between the frequency range of **20–20kHz**

```python
# Test the unaltered sound of "sound4.wav"
wave4.make_audio()
```

Select the audio file and isolate the part with the high frequency using a **high-pass filter**

```python
spectrum4 = wave4.make_spectrum(full=True)
spectrum4.high_pass(10000)
spectrum4.plot()
decorate(xlabel='Frequency (Hz)')
```

Using High pass filter we are able to isolate any ultrasonic sounds from 20kHz onwards as shown below. We can hear some mumbling just not very clear

```python
isolated_wave = spectrum4.make_wave()
isolated_wave.normalize()
isolated_wave.make_audio()
```

To demodulate the AM signal we need to modulate it by a "carrier wave", which is a cosine, **10 kHz at the begining using isolated wave** as input.

### Testing with 10kHz

```python
#Testing with a 10kHz frequency
carrier_sig = thinkdsp.CosSignal(freq=10000)
carrier_wave = carrier_sig.make_wave(duration=isolated_wave.duration, framerate=isolated_wave.framerate)
```

Demodulate by * the carrier signal with AM isolated signal

```python
demodulated = isolated_wave * carrier_wave
```

This is the Demodulated signal. As we can see the each half of the spectrum gets split and shifted towards 0. However in this frequency of 10kHz, the havles seems too far and I have undershot the carrier frequency.

```python
demodulated_spectrum = demodulated.make_spectrum(full=True)
demodulated_spectrum.plot()
```

This is what it sounds like when the carrier frequency is **wrong**

```python
demodulated.make_audio()
```

### Testing 19kHz after trail and error 

```python
#Testing with a 19kHz frequency
carrier_sig1 = thinkdsp.CosSignal(freq=19000)
carrier_wave1 = carrier_sig1.make_wave(duration=isolated_wave.duration, framerate=isolated_wave.framerate)
```

Demodulate by * the carrier signal with AM isolated signal

```python
demodulated1 = isolated_wave * carrier_wave1
```

This is the Demodulated signal. As we can see the Each half of the spectrum gets split and shifted towards 0. In this carrier frequency of **19kHz** , the havles gets shifted closer to 0 and are added up.

### This is the desired demodulation

```python
Newdemodulated_spectrum = demodulated1.make_spectrum(full=True)
Newdemodulated_spectrum.plot()
```

This is what it sounds like when the carrier frequency is **correct**, However there are still high frequency noises at the background

```python
demodulated1.make_audio()
```

By Applying a low-pass filter, we can filter out the high frequency pitches and get a cleaner sound 

```python
#Low-pass cut off freq above 8000Hz
Newdemodulated_spectrum.low_pass(8000)
Newdemodulated_spectrum.plot()
```

## The audio is clearer and we can hear the secret code"1891"

```python
filtered = Newdemodulated_spectrum.make_wave()
filtered.unbias()
filtered.normalize()
filtered.make_audio()
```
