```python
# Wave package for reading and writing .wav audio files
import wave
#import libraries for audio analysis
import thinkdsp
from thinkdsp import read_wave
```

Read selected audio file

```python
# Read ONLY selected audio file
# 'rb' = Read only mode
mySound = wave.open("audiofiles/Ex3_sound5.wav",mode = 'rb')
```

### Encode function- LSB Algorithm
#### similar to exercise 3.1 and create frameByte

```python
# Read frames and convert to byte array
frameBytes = bytearray(list(mySound.readframes(mySound.getnframes())))
```

The "secret" message as string variable

```python
string = "Father Christmas does not exist"
print(string)
```

**Append dummy data** to fill out rest of the bytes. Decoder shall detect and remove these characters

```python
string = string + int((len(frameBytes)-(len(string)*8*8))/8) *'#'
##  Adds hash # at the back of string
```

Convert the secret text to bits array

```python
bits = list(map(int, ''.join([bin(ord(i)).lstrip('0b').rjust(8,'0') for i in string])))
```

Replace LSB of audio data bit by bit from the text bit array

The **enumerate()** method adds a counter to an iterable and return bits

```python
for i, bit in enumerate(bits):
    frameBytes[i] = (frameBytes[i] & 254) | bit
# Get the modified bytes
frame_modified = bytes(frameBytes)
```

Write bytes to a new wave audio file **"Encoded.wav"**
'wb' = Write only mode.

```python
# Write bytes to a new wave audio file
with wave.open('EncodedAudio.wav', 'wb') as fd:
    fd.setparams(mySound.getparams())
    fd.writeframes(frame_modified)
mySound.close()
```

A new audio file called **"EncodedAudio.wav"** is created and contains the hidden message 

### Extraction function 
#### similar decoding method as exercise 3.1

```python
# Read selected audio file
DecodeWave = wave.open("EncodedAudio.wav",mode = 'rb')
wave = read_wave("EncodedAudio.wav")
#Play the selected audio
wave.make_audio()
```

```python
# Read frames and convert to byte array
frameBytes = bytearray(list(DecodeWave.readframes(DecodeWave.getnframes())))
print ("Number of frames: ",DecodeWave.getnframes())
# Display first 10 frames of the list
print ("First 10 of frames list: ",frameBytes[0:10])
```

```python
#Extract the least Significant bit of each byte
extracted = [frameBytes[i] & 1 for i in range(len(frameBytes))]
```

```python
#Convert the byte array back to string
string = "".join(chr(int("".join(map(str,extracted[i:i+8])),2))for i in range(0,len(extracted),8))
#Display the first 100 string characters since the end result is too huge 
print(string[0:100])
```

```python
# Remove any filler characters
decoded = string.split("###")[0]
```

```python
# Print the decoded TEXT
print("The hidden message is: "+decoded)
DecodeWave.close()
```

#### The secret message is successfully decoded !!
